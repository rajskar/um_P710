```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------
# 1. GENERATE DATA
# -----------------
# We'll create some sample data for a simple linear relationship
# with some random noise to make it more realistic.
np.random.seed(0)
X = np.arange(1, 10, 0.5) # A range of x values
y = 2 * X + 1 + np.random.randn(len(X)) * 2 # y = 2x + 1 with some noise

# -----------------------------------------------
# 2. PERFORM LINEAR REGRESSION (NO OUTLIER) - VECTOR FORM
# -----------------------------------------------
# The Normal Equation in vector form is: theta = (X^T * X)^-1 * X^T * y
# where X is the design matrix and y is the target vector.

# Create the design matrix for the initial data.
# The first column is a vector of ones for the bias (y-intercept).
# The second column is the feature data.
X_matrix_initial = np.vstack([np.ones(len(X)), X]).T

# Calculate the parameters (theta) using the normal equation.
# np.linalg.inv() calculates the inverse of a matrix.
theta_initial = np.linalg.inv(X_matrix_initial.T @ X_matrix_initial) @ X_matrix_initial.T @ y

# Extract the slope (m) and y-intercept (b) from the theta vector.
b_initial, m_initial = theta_initial

print(f"Initial Slope (m) without outlier: {m_initial:.2f}")
print(f"Initial Y-Intercept (b) without outlier: {b_initial:.2f}")

# -------------------------------------
# 3. ADD AN OUTLIER TO THE DATA
# -------------------------------------
# Add a single data point that is far from the linear trend.
# This point will significantly influence the regression line.
X_outlier_data = np.append(X, 9.5)
y_outlier_data = np.append(y, -5) # The outlier is a very low value

# ------------------------------------------------
# 4. PERFORM LINEAR REGRESSION (WITH OUTLIER) - VECTOR FORM
# ------------------------------------------------
# Create the design matrix for the data with the outlier.
X_outlier_matrix = np.vstack([np.ones(len(X_outlier_data)), X_outlier_data]).T

# Re-calculate the parameters (theta) with the outlier included.
theta_outlier = np.linalg.inv(X_outlier_matrix.T @ X_outlier_matrix) @ X_outlier_matrix.T @ y_outlier_data

b_outlier, m_outlier = theta_outlier

print("\n------------------------------")
print("Parameters with Outlier Added:")
print(f"New Slope (m): {m_outlier:.2f}")
print(f"New Y-Intercept (b): {b_outlier:.2f}")
print("------------------------------")

# ----------------------
# 5. PREDICTED DATA
# ----------------------
# Now, let's use our calculated `theta_outlier` to generate
# the predicted line. The prediction is a simple matrix multiplication.
y_predicted = X_outlier_matrix @ theta_outlier

# --------------------------
# 6. PLOT THE DATA
# --------------------------
# Create a new figure and axes for the plot
plt.figure(figsize=(10, 6))

# Plot the original data as a scatter plot
plt.scatter(X, y, color='blue', label='Original Data')

# Plot the outlier with a distinct marker and color
plt.scatter(X_outlier_data[-1], y_outlier_data[-1], color='red', marker='X', s=100, label='Outlier')

# Plot the new regression line on the same axes
plt.plot(X_outlier_data, y_predicted, color='red', linewidth=2, label='Predicted Regression Line (with Outlier)')

# Add labels, title, and a legend for better readability
plt.title('Linear Regression with an Outlier')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.grid(True) # Add a grid for easier reading

# Display the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------
# 1. GENERATE DATA
# -----------------
# We'll create some sample data for a simple linear relationship
# with some random noise to make it more realistic.
np.random.seed(0)
X = np.arange(1, 10, 0.5) # A range of x values
y = 2 * X + 1 + np.random.randn(len(X)) * 2 # y = 2x + 1 with some noise

# -----------------------------------------------------------------
# 2. PERFORM LINEAR REGRESSION (NO OUTLIER) - ITERATIVE GRADIENT DESCENT
# -----------------------------------------------------------------
# We will use gradient descent to find the optimal parameters (theta).
# The algorithm iteratively updates parameters to minimize the cost function.

# Configuration for Gradient Descent
learning_rate = 0.01  # How big of a step to take with each iteration
n_iterations = 1000   # The number of times to update the parameters

# Create the design matrix for the initial data.
# The first column is a vector of ones for the bias (y-intercept).
X_matrix_initial = np.vstack([np.ones(len(X)), X]).T
n_initial = len(X)

# Initialize parameters (theta) to zeros.
theta_initial = np.zeros(X_matrix_initial.shape[1])

# Create a list to store the loss at each iteration
loss_history_initial = []

# Perform gradient descent
for _ in range(n_iterations):
    # Calculate the predicted values (hypothesis)
    predictions = X_matrix_initial @ theta_initial
    # Calculate the error
    error = predictions - y
    # Calculate the mean absolute error and store it
    cost = np.mean(np.abs(error))
    loss_history_initial.append(cost)
    # Calculate the gradient (partial derivative of the cost function)
    gradient = (X_matrix_initial.T @ error) / n_initial
    # Update the parameters
    theta_initial = theta_initial - learning_rate * gradient

# Extract the slope (m) and y-intercept (b) from the final theta vector.
b_initial, m_initial = theta_initial

print("-----------------------------------------")
print("Parameters from model without the outlier")
print(f"Initial Slope (m): {m_initial:.2f}")
print(f"Initial Y-Intercept (b): {b_initial:.2f}")
print("-----------------------------------------")

# Calculate the predicted y-values for the original line
y_predicted_initial = X_matrix_initial @ theta_initial

# -------------------------------------
# 3. ADD AN OUTLIER TO THE DATA
# -------------------------------------
# Add a single data point that is far from the linear trend.
# This point will significantly influence the regression line.
X_outlier_data = np.append(X, 9.5)
y_outlier_data = np.append(y, -5) # The outlier is a very low value

# -----------------------------------------------------------------
# 4. PERFORM LINEAR REGRESSION (WITH OUTLIER) - ITERATIVE GRADIENT DESCENT
# -----------------------------------------------------------------
# Re-run gradient descent with the outlier data included.

# Create the design matrix for the data with the outlier.
X_outlier_matrix = np.vstack([np.ones(len(X_outlier_data)), X_outlier_data]).T
n_outlier = len(X_outlier_data)

# Initialize parameters (theta) to zeros for the new data.
theta_outlier = np.zeros(X_outlier_matrix.shape[1])

# Create a list to store the loss at each iteration
loss_history_outlier = []

# Perform gradient descent
for _ in range(n_iterations):
    predictions = X_outlier_matrix @ theta_outlier
    error = predictions - y_outlier_data
    # Calculate the mean absolute error and store it
    cost = np.mean(np.abs(error))
    loss_history_outlier.append(cost)
    gradient = (X_outlier_matrix.T @ error) / n_outlier
    theta_outlier = theta_outlier - learning_rate * gradient

b_outlier, m_outlier = theta_outlier

print("\n--------------------------------------")
print("Parameters from model with the outlier")
print(f"New Slope (m): {m_outlier:.2f}")
print(f"New Y-Intercept (b): {b_outlier:.2f}")
print("--------------------------------------")

# ----------------------
# 5. PREDICTED DATA
# ----------------------
# Now, let's use our calculated `theta_outlier` to generate
# the predicted line.
y_predicted_outlier = X_outlier_matrix @ theta_outlier

# --------------------------
# 6. PLOT THE DATA AND REGRESSION LINES
# --------------------------
# Create a new figure and axes for the plot
plt.figure(figsize=(10, 6))

# Plot the original data as a scatter plot
plt.scatter(X, y, color='blue', label='Original Data')

# Plot the outlier with a distinct marker and color
plt.scatter(X_outlier_data[-1], y_outlier_data[-1], color='red', marker='X', s=100, label='Outlier')

# Plot the original regression line (without outlier)
plt.plot(X, y_predicted_initial, color='green', linestyle='--', linewidth=2, label='Regression Line (without Outlier)')

# Plot the new regression line (with outlier)
plt.plot(X_outlier_data, y_predicted_outlier, color='red', linewidth=2, label='Regression Line (with Outlier)')

# Add labels, title, and a legend for better readability
plt.title('Linear Regression with and without an Outlier')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.grid(True) # Add a grid for easier reading

# --------------------------
# 7. PLOT THE LOSS HISTORY
# --------------------------
# Create a separate figure for the loss history
plt.figure(figsize=(10, 6))

# Plot the loss history for both models
plt.plot(range(n_iterations), loss_history_initial, color='green', label='Loss without Outlier')
plt.plot(range(n_iterations), loss_history_outlier, color='red', label='Loss with Outlier')

# Add labels, title, and a legend
plt.title('Loss per Iteration for Gradient Descent')
plt.xlabel('Number of Iterations')
plt.ylabel('Mean Absolute Error (MAE) Loss')
plt.legend()
plt.grid(True)

# Display both plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------
# 1. GENERATE DATA
# -----------------
# We'll create some sample data for a simple linear relationship
# with some random noise to make it more realistic.
np.random.seed(0)
X = np.arange(1, 10, 0.5) # A range of x values
y = 2 * X + 1 + np.random.randn(len(X)) * 2 # y = 2x + 1 with some noise

# -----------------------------------------------------------------
# 2. PERFORM LINEAR REGRESSION (NO OUTLIER) - ITERATIVE GRADIENT DESCENT
# -----------------------------------------------------------------
# We will use gradient descent to find the optimal parameters (theta).
# The algorithm iteratively updates parameters to minimize the cost function.

# Configuration for Gradient Descent
learning_rate = 0.01  # How big of a step to take with each iteration
n_iterations = 1000   # The number of times to update the parameters

# Create the design matrix for the initial data.
# The first column is a vector of ones for the bias (y-intercept).
X_matrix_initial = np.vstack([np.ones(len(X)), X]).T
n_initial = len(X)

# Initialize parameters (theta) to zeros.
theta_initial = np.zeros(X_matrix_initial.shape[1])

# Create a list to store the loss at each iteration
loss_history_initial = []

# Perform gradient descent (using MAE loss)
for _ in range(n_iterations):
    # Calculate the predicted values (hypothesis)
    predictions = X_matrix_initial @ theta_initial
    # Calculate the error
    error = predictions - y
    # Calculate the mean absolute error and store it
    cost = np.mean(np.abs(error))
    loss_history_initial.append(cost)
    # Calculate the gradient (partial derivative of the MAE cost function)
    gradient = (X_matrix_initial.T @ np.sign(error)) / n_initial
    # Update the parameters
    theta_initial = theta_initial - learning_rate * gradient

# Extract the slope (m) and y-intercept (b) from the final theta vector.
b_initial, m_initial = theta_initial

print("-----------------------------------------")
print("Parameters from model without the outlier")
print(f"Initial Slope (m): {m_initial:.2f}")
print(f"Initial Y-Intercept (b): {b_initial:.2f}")
print("-----------------------------------------")

# Calculate the predicted y-values for the original line
y_predicted_initial = X_matrix_initial @ theta_initial

# -------------------------------------
# 3. ADD AN OUTLIER TO THE DATA
# -------------------------------------
# Add a single data point that is far from the linear trend.
X_outlier_data = np.append(X, 9.5)
y_outlier_data = np.append(y, -15) # The outlier is a very low value

# -----------------------------------------------------------------
# 4. PERFORM LINEAR REGRESSION (WITH OUTLIER) - MAE LOSS
# -----------------------------------------------------------------
# Re-run gradient descent with the outlier data included, using MAE.

# Create the design matrix for the data with the outlier.
X_outlier_matrix = np.vstack([np.ones(len(X_outlier_data)), X_outlier_data]).T
n_outlier = len(X_outlier_data)

# Initialize parameters (theta) to zeros for the new data.
theta_mae = np.zeros(X_outlier_matrix.shape[1])

# Create a list to store the MAE loss
loss_history_mae = []

# Perform gradient descent (using MAE loss)
for _ in range(n_iterations):
    predictions = X_outlier_matrix @ theta_mae
    error = predictions - y_outlier_data
    cost = np.mean(np.abs(error))
    loss_history_mae.append(cost)
    gradient = (X_outlier_matrix.T @ np.sign(error)) / n_outlier
    theta_mae = theta_mae - learning_rate * gradient

b_mae, m_mae = theta_mae

print("\n--------------------------------------")
print("Parameters from model with the outlier (MAE)")
print(f"New Slope (m): {m_mae:.2f}")
print(f"New Y-Intercept (b): {b_mae:.2f}")
print("--------------------------------------")

# Calculate the predicted y-values for the MAE line
y_predicted_mae = X_outlier_matrix @ theta_mae

# -----------------------------------------------------------------
# 5. PERFORM LINEAR REGRESSION (WITH OUTLIER) - MSE LOSS
# -----------------------------------------------------------------
# Re-run gradient descent with the outlier data included, using MSE.

# Initialize parameters (theta) to zeros for the new data.
theta_mse = np.zeros(X_outlier_matrix.shape[1])

# Create a list to store the MSE loss
loss_history_mse = []

# Perform gradient descent (using MSE loss)
for _ in range(n_iterations):
    predictions = X_outlier_matrix @ theta_mse
    error = predictions - y_outlier_data
    cost = np.mean(error**2) # MSE loss function
    loss_history_mse.append(cost)
    # Calculate the gradient for MSE loss
    gradient = (X_outlier_matrix.T @ error) * 2 / n_outlier
    theta_mse = theta_mse - learning_rate * gradient

b_mse, m_mse = theta_mse

print("\n--------------------------------------")
print("Parameters from model with the outlier (MSE)")
print(f"New Slope (m): {m_mse:.2f}")
print(f"New Y-Intercept (b): {b_mse:.2f}")
print("--------------------------------------")

# Calculate the predicted y-values for the MSE line
y_predicted_mse = X_outlier_matrix @ theta_mse

# --------------------------
# 6. PLOT THE DATA AND REGRESSION LINES
# --------------------------
# Create a new figure and axes for the plot
plt.figure(figsize=(10, 6))

# Plot the original data as a scatter plot
plt.scatter(X, y, color='blue', label='Original Data')

# Plot the outlier with a distinct marker and color
plt.scatter(X_outlier_data[-1], y_outlier_data[-1], color='red', marker='X', s=100, label='Outlier')

# Plot the original regression line (without outlier)
plt.plot(X, y_predicted_initial, color='green', linestyle='--', linewidth=2, label='Regression Line (without Outlier)')

# Plot the new regression line (with outlier) using MAE
plt.plot(X_outlier_data, y_predicted_mae, color='red', linewidth=2, label='Regression Line (with Outlier - MAE)')

# Plot the new regression line (with outlier) using MSE
plt.plot(X_outlier_data, y_predicted_mse, color='purple', linestyle=':', linewidth=2, label='Regression Line (with Outlier - MSE)')

# Add labels, title, and a legend for better readability
plt.title('Linear Regression with MAE and MSE Loss')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.legend()
plt.grid(True) # Add a grid for easier reading

# --------------------------
# 7. PLOT THE LOSS HISTORY
# --------------------------
# Create a separate figure for the loss history
plt.figure(figsize=(10, 6))

# Plot the loss history for all three models
plt.plot(range(n_iterations), loss_history_initial, color='green', label='Loss without Outlier (MAE)')
plt.plot(range(n_iterations), loss_history_mae, color='red', label='Loss with Outlier (MAE)')
plt.plot(range(n_iterations), loss_history_mse, color='purple', label='Loss with Outlier (MSE)')

# Add labels, title, and a legend
plt.title('Loss per Iteration for Gradient Descent')
plt.xlabel('Number of Iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Display both plots
plt.show()
```
