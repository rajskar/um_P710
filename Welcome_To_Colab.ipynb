````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````

````
import numpy as np
import matplotlib.pyplot as plt

# Generate 2D normally distributed samples
mean = [2, 2]
cov = [[1, -0.8], [-0.8, 1]] # identity covariance matrix
num_samples = 100

samples = np.random.multivariate_normal(mean, cov, num_samples)

# Display the samples
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('2D Normally Distributed Samples')

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
````

````
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Generate the first dataset
mean1 = [2, 2]
cov1 = [[1, 0.8], [0.8, 1]] # Covariance matrix for the first dataset
num_samples1 = 500

samples1 = np.random.multivariate_normal(mean1, cov1, num_samples1)

# Generate the second dataset
mean2 = [4, 4]
cov2 = [[1, 0.8], [0.8, 1]] # Covariance matrix for the second dataset
num_samples2 = 500

samples2 = np.random.multivariate_normal(mean2, cov2, num_samples2)

# Visualize the datasets
plt.figure(figsize=(8, 6))
plt.scatter(samples1[:, 0], samples1[:, 1], alpha=0.5, label='Dataset 1 (Mean: [2, 2])')
plt.scatter(samples2[:, 0], samples2[:, 1], alpha=0.5, label='Dataset 2 (Mean: [4, 4])')

# Create a grid for the contour plots
x, y = np.mgrid[0:6:.1, 0:6:.1]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x
pos[:, :, 1] = y

# Calculate the density of the first distribution
rv1 = multivariate_normal(mean1, cov1)
plt.contour(x, y, rv1.pdf(pos), cmap='Blues', alpha=0.5)

# Calculate the density of the second distribution
rv2 = multivariate_normal(mean2, cov2)
plt.contour(x, y, rv2.pdf(pos), cmap='Reds', alpha=0.5)

plt.title('Two Datasets of 2D Normally Distributed Samples with Contours')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.show()
````
